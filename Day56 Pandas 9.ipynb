```python
import pandas as pd
```

```python
 # Slicing data from the Pandas DataFrame:
 # 1) loc for label-based slicing:Selecting rows and columns using labels # last index included
 # 2) iloc for positional slicing:Selecting rows and columns by integer position # last index excluded
 # 3) at for quick access to single values
 # 4) iat for quick access to single values by position
```

```python
data2=pd.read_excel(r"C:\Ganesh\Minskole Python Batch\input data.xlsx")
```

```python
data2
```

```python
# loc
```

```python
data2.loc[0:3 , ["First Name","Last Name"]] # loc :label-based slicing:Selecting rows and columns using labels
                                            # last index included
```

```python
# iloc
```

```python
data2.iloc[0:3 , ["First Name","Last Name"]]
```

```python
data2.iloc[0:3,[0,1]]  # positional slicing:Selecting rows and columns by integer position
                        # last index excluded
```

```python
data2.loc[0:3 , ["First Name","Last Name"]] # loc :label-based slicing:Selecting rows and columns using labels
                                            # last index included
```

```python
data2.loc[0]
```

```python
data2
```

```python
# at: Accessing a single value
data2.at[3, 'First Name']
       # Row labelled index  Column labelled index
```

```python
data2.at[1:4, 'First Name']
```

```python
# iat: Accessing a single value by position
data2.iat[1, 2]  
# Row index , Column index
```

```python
data2.head()
```

```python
# # Rename 
# # Syntax
# 1)Row labelled
#  dataframe_name.rename(index         = {1:"One",               4:"Four"} , inplace= True/False)
#                        For Row         Old value : New value,

# 2)Columns labelled
#  dataframe_name.rename(columns         = {1:"One",               4:"Four"} , inplace= True/False)
#                        For Column         Old value : New value,
```

```python
data2.rename(index={1:"One",4:"Four"},inplace=True)
```
