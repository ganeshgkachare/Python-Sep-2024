```python
import pandas as pd
```

```python
# Read excel file in pandas
# Syntax
# pd.read_excel("file_path/file_name.extension") ==> Excel
# pd.read_csv("file_path/file_name.csv")  ==> Csv
```

```python
# path
# extensions ==> .xlsx = excel file  
                 # .csv= csv file
# single / double slash / raw string

# sheet_name : Specifies which sheet in the Excel file to read. By default read first sheet
# header (None/other number) : Define which row to use as the header (column names).
# usecols : Define which columns to read from the Excel sheet.
# skiprows : Skips a specified number of rows at the start
```
