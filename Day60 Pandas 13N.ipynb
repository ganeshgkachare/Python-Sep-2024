```python
# single column same name
data1=pd.merge(Catalog,Eng,on="Roll No")
```

```python
# Multiple column same name
data2=pd.merge(data1,Phy,on=["First Name","Last Name"])
```

```python
# Different column Different name
```

```python
data3=pd.merge(data2,Chem,left_on=["First Name","Last Name"],right_on=["FName","LName"])
```

```python
data3.head(2)
```

```python
Math.head(2)
```

```python
# Different column same name
```

```python
data4=pd.merge(data3,Math,left_on="Roll No",right_on="ID") #by default how="inner"
```

```python
pd.merge(data3,Math,left_on="Roll No",right_on="ID") #by default how="inner" ==> common records present in output
```

```python
pd.merge(data3,Math,left_on="Roll No",right_on="ID",how="left") #how="left" ==> Base data if left dataframe (data3) [All records from left df present output]+ Matched record from right df present in output
```
