```python
# pivot 

# Syntax:
# DataFrame.pivot(index=None, columns=None, values=None)

# index: The column(s) to set as the new index of the DataFrame.
# columns: The column whose unique values will become the new columns.
# values: The column(s) to populate the values in the reshaped DataFrame.

# Notes:
# The pivot() method expects the data should be unique. If you have duplicate entries, you will get a ValueError.
# In that case, consider using pivot_table().
```

```python
import pandas as pd
```

```python
data3=pd.read_excel(r"C:\Ganesh\Minskole Python Batch\Reshape Transform data.xlsx",sheet_name="Unique")
```

```python
data4=pd.read_excel(r"C:\Ganesh\Minskole Python Batch\Reshape Transform data.xlsx",sheet_name="Duplicate")
```

```python
data3.head()
```

```python
data3.pivot(index=["First Name","Last Name"],columns="Subject",values="Marks")
```

```python
data3.pivot(index=["First Name","Last Name"],columns="Subject",values="Marks").reset_index()
```

```python
data33=data3.pivot(index=["First Name","Last Name"],columns="Subject",values="Marks").reset_index()
```

```python
data33
```

```python
data4.pivot(index=["First Name","Last Name"],columns="Subject",values="Marks").reset_index()
```

```python
# pivot_table

# Syntax:
# DataFrame.pivot_table(values=None, index=None, columns=None, aggfunc='mean', fill_value=None)

# values: The column(s) to aggregate.
# index: The column(s) to group by.
# columns: The column whose unique values will become the new columns.
# aggfunc: The aggregation function to apply (default is mean).
# fill_value: Value to replace missing values with.


# The pivot_table() method is an enhanced version of pivot().
# Handle duplicate values by applying aggregation function (like sum, mean, etc.) in case of duplicate entries.
```

```python
data3.pivot(index=["First Name","Last Name"],columns="Subject",values="Marks").reset_index()
```

```python
data4.pivot_table(index=["First Name","Last Name"],columns="Subject",values="Marks",aggfunc="mean").reset_index()
```

```python
data4.pivot_table(index=["First Name","Last Name"],columns="Subject",values="Marks",aggfunc="sum")
```

```python
data3.pivot_table(index=["First Name","Last Name"],columns="Subject",aggfunc="mean")
```

```python
#  melt

# Syntax:
# DataFrame.melt(id_vars=None, value_vars=None, var_name=None, value_name=None)

# id_vars: The columns to keep as identifier variables.
# value_vars: The columns to unpivot (melt).
# var_name: The name of the new column that will contain the original column names.
# value_name: The name of the new column that will contain the values.

# The melt() method is used to "unpivot" data . It is essentially the inverse of the pivot() function.
```

```python
data33
```

```python
data33.melt(id_vars=["First Name","Last Name"],value_vars=["English","Physics","Chemistry","Mathematics"],
           var_name="Sub",value_name="Marks")
```

```python
data33.melt(id_vars=["First Name","Last Name"],value_vars=["English","Physics"],
           var_name="Sub",value_name="Marks")
```
